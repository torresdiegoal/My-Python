Generate a funcion to check if a string is palindrome or not.

Example of palindrome word: "Solos"

```
# Add your function here

def is_palindrome(word):
  #Check if a word is a Palindrome
  return word == word[::-1]
```

## Exercise 1
Using pandas
Retrieve from df dataframe the number of cases and deaths in South America Countries.  (Hint: Use groupby by continent and location)

```
# Add your solution here
South_America_df=df[df['continent']=='South America']
#Select my columns and group by their location
group=South_America_df[['location','new_cases','new_deaths','continent']].groupby('continent', 'location')
result=group.agg(total_cases=('new_cases','sum'),
                  total_deaths=('new_deaths','sum')
                  ).reset.index()
```

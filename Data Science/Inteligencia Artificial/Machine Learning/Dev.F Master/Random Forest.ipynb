```python
 ### El mÃ©todo random forest selecciona de manera aleatoria observaciones del df, y de igual manera, deja deliberadamente fuera
  #de la base de datos otras. 
        # arboles de clasif o regres pero pluralizado en cientos o miles
        # los resultados de sus modelos permiten obtener:
            # la media en caso de variab numerics
            # la categoria con mas casos

from sklearn import tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

inpath = "D:/Disco D/Big Data/Data/"
df = pd.read_csv(inpath + "base_datos_2008.csv")
df = df.dropna(subset = ["ArrDelay"]) 
df = df.sample(frac=1)
df = df.head(100000)
df_test = df.tail(100000)
df
```

```python
clf =tree.DecisionTreeClassifier() #otros param max_features = nÂ°, bootstrap = boolean --> realiza muestreo con reemplazo
                                                                    # generando muestras aleatorias de las filas para rehacer los 
                                                                    # bosques

x = df[["AirTime", "Distance", "TaxiIn", "TaxiOut", "DepTime", "DepDelay"]]
x_test = df_test[["AirTime", "Distance", "TaxiIn", "TaxiOut", "DepTime", "DepDelay"]]
y = df["ArrDelay"] > 10
y_test = df_test["ArrDelay"] > 10 # variable respuesta, booleana

clf = clf.fit(x,y)
y_pred_test = clf.predict(x_test)
np.mean(y_test == y_pred_test)
```

```python
clf = RandomForestClassifier(n_estimators = 100) #esta generando 100 arboles
clf.fit(x,y)
y_pred_test = clf.predict(x_test)

clf.feature_importances_ # nos da la reelevancia de las variables usadas, em el caso de abajo, la mas relevante es Depdelay
```

```python
np.mean(y_test == y_pred_test)
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inpath = "D:/Disco D/bigData/data/"
df = pd.read_csv(inpath + "base_datos_2008.csv", nrows = 1e5)
newdf = df[["AirTime", "DepDelay"]].dropna()
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0).fit(newdf)
kmeans.labels_ #permite ver la categoria del grupo al que pertenece cada dato
len(kmeans.labels_)
```

```python
np.unique(kmeans.labels_, return_counts = True) # conteo de datos por cada categoria
```

```python
plt.scatter(newdf["AirTime"], newdf["DepDelay"], c = kmeans.labels_)
```

```python
kmeans.cluster_centers_  # nos da el centroide de cada clasificacion
```

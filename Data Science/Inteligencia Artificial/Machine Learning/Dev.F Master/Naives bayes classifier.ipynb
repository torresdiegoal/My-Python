```python
 ### PROBABILIDAD CONDICIONAL!
    # naive debido a la "ingenuidad" de creer que todas las variales incluidas dentro del modelo son independientes, y no es así
    # El teorema de Bayes permite detectar causalidades entre eventos
    
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inpath = "D:/Disco D/bigData/data/"
df = pd.read_csv(inpath + "base_datos_2008.csv")
df = df.sample(frac=1)
df = df.head(500000)
df = df.dropna(subset = ["ArrDelay"]) #elimina las filas con NA's solamente de la columna ArrDelay
```

```python
y = df["ArrDelay"] > 0
```

```python
df["Month"] = df["Month"].apply(str) # CONVERTIR A STRING
df["DayofMonth"] = df["DayofMonth"].apply(str) 
df["DayOfWeek"] = df["DayOfWeek"].apply(str) 
df["TailNum"] = df["TailNum"].apply(str) 

x = pd.get_dummies(df[["Month", "DayofMonth", "DayOfWeek", "Origin", "Dest", "UniqueCarrier"]])
```

```python
x.head()
```

```python
clf = BernoulliNB() ## Al ser un modelo binario, su poder de predicción es corto
clf.fit(x, y)
y_pred = clf.predict(x)
```

```python
np.mean(y_pred == y) # Porcentaje de prediccion correcta (y_pred) con respecto a los datos reales de y
```

```python
1-np.mean(y) # Si se compara con la media de los datos reales, solo estamos ganando un 3% de poder de prediccion, muy poco
                                                                                                    # para ser un modelo
                                # Lo anterior sucede porque manejamos variables poco explicativas y es un modelo más usado para el
                                # analisis textual
```

```python
x = df[["AirTime", "Distance", "TaxiIn", "TaxiOut"]] # "DepDelay" variable más correlacionada a ArrDelay
clf = GaussianNB()
clf.fit(x, y)
y_pred = clf.predict(x)
```

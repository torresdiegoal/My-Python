```python
##### MUY COSTOSOS COMPUTACIONALMENTE ####

from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inpath = "D:/Disco D/bigData/data/"
df = pd.read_csv(inpath + "base_datos_2008.csv", nrows = 1e4)
newdf = df[["AirTime", "DepDelay"]].dropna()
```

```python
newdf = df[["DepDelay", "ArrDelay", "Distance", "AirTime", "Origin"]].dropna() # Correr de nueov el codigo con mas variables como aqui
```

```python
clstr = AgglomerativeClustering(n_clusters = 5)
clstr.fit(newdf)

# affinity = "" / euclidean, L1, L2, manhattan, cosine, precomputed
# ej, if affinity "L1", linkage = "complete"
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"], c = clstr.fit_predict(newdf))
```

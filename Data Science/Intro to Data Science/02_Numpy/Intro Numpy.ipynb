```python
 ### Test Chi Cuadrado
    # Importante porque permite encontrar relaciones entre variables cualitativas mediante tablas de contingencia
    #1  
    
df = pd.read_csv(inpath + "base_datos_2008.csv")
df
```

```python
np.random.seed(0)
df = df[df["Origin"].isin(["HOU", "ATL", "IND"])]
df = df.sample(frac = 1) #ordena a partir del orden de las primeras columnas
df = df[0:100000]  # se selecciona un pequeño grupo porque a mayor escala los patrones tienden a repetirse
df
```

```python
 # haremos una tabla de frecuencias con la clasificacion booleana a partir de la siguiente condicion
    
df["BigDelay"] = df["ArrDelay"] > 30
observados = pd.crosstab(index = df["BigDelay"], columns = df["Origin"], margins = True) # Tabla de contingencia #margins = los totales
observados
```

```python
 # Importamos el paquete del test de chi
    
from scipy.stats import chi2_contingency
test = chi2_contingency(observados) # arroja tres resultados a) estadistico, b) p-valor, c) matriz de esperados si no hubiera relacion entre variables
test
```

```python
esperados = pd.DataFrame(test[3])
esperados
```

```python
 # los anteriores valores son aproximaciones teoricas, pero para su mejor intepretacion, se trabajara con los valores relativos
    
esperados_relatv = round(esperados.apply(lambda r:r/len(df) * 100, axis = 1), 2) # lambda porque es funcion sin nombre
esperados_relatv
```

```python
 # lo mismo para la tabla de observados
    
observados_relatv = round(observados.apply(lambda r:r/len(df) * 100, axis = 1), 2) # lambda porque es funcion sin nombre
observados_relatv
```

```python
test[1]
```

```python
 # Resumen Test de Chi Squared
    
# Si p-valor < 0.05, hay diferencias significativas, hay relacion entre variables
# Si p-valor > 0.05, no hay diferencias significativas, No habria relacion entre variables
```

```python
 ### Analisis de Outliers
    
from sklearn.covariance import EllipticEnvelope #para determinar outliers en varias columnas a la vez

outliers = EllipticEnvelope(contamination = 0.01) #selecciona el 1% de los datos mas alejados de la media
var_list = ["DepDelay", "TaxiIn", "TaxiOut", "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]
x = np.array(df.loc[:,var_list].dropna()) # accede a filas o columnas por etiquetas o booleanos con df.loc
len(x)
```

```python
# entrenar el modelo Elliptic

outliers.fit(x) # lee arrays unicamente
pred = outliers.predict(x)
elipse_outl = np.where(pred == -1) # nos muestra los indices en donde estan los valores filtrados
elipse_outl
```

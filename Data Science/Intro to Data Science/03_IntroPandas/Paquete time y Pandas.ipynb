```python
 ### groupby
    
#1 agrupa por maximo de los retrasos de llegada por cada dia de la semana 
df.groupby("DayOfWeek")["ArrDelay"].max() #o mean(), min(), sd(), numero de casos count(), o resumen estadistico describe() 

dfATLHOU.groupby(["DayOfWeek", "Origin"])["ArrDelay"].mean() # Agrupacion por mas de una categoria

#2 Es posible guardar el groupby en un objeto y aplicar la funcion posteriormente
mygb = dfATLHOU.groupby(["DayOfWeek", "Origin"])["ArrDelay"]
mygb.describe()
```
